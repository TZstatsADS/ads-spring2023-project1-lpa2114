## How positive or negative do these schools perceive religion and religion-adjacent concepts?

To hone in on schools of interest, we will only look at a subset of the options for schools of thought. Specifically, I have selected Feminism, Capitalism, Communism, Nietzshe, Analytic Philosophy, and Continental Philosophy. I chose the first three because they are the most well-known and relevant schools among the modern population, especially those who are not familiar with philosophy. Nietzshe is very famously known as an atheist and preaches as such, so it would be good to see how this assessment views Nietzshe in sentiment relative to the other schools. Finally, Analytical and Continental Philosophies are often pitted against one another and immediately conflict, so there is the interest in seeing the contrast between their perspectives on religion. 

```python
import nltk
```

```python
nltk.download(['stopwords','vader_lexicon']) #download aspects of nltk needed for cleaning and analysis
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
```

In order to analyze the most commonly used words around religion for each school, we remove stop words, or common words and characters which don't alter meaning in a sentence (e.g. at, to) and turn the table of sentences into a long list of all of the words in the sentences in order. 

```python
stop_words = set(stopwords.words('english')+['-pron-',',','.',';',':','?','!','-','--'])
```

```python
def concat_clean_words(school): #returns list of words from specified school in order
    cond = phi_data.loc[:,'school'] == school
    text = ' '.join(phi_data.loc[cond,'lemmatized_str']) #lemmatized for word cloud later
    text_tokens = text.split()
    return [word.lower() for word in text_tokens if word.lower() not in stop_words]
```

Using the function to retrieve that long list of words, now we create a function to apply sentiment analysis on that list and retrieve an average sentiment for words surrounding any specified topic. In this case, we are most interested in religion and God. God is useful for this case as Christianity is by far the more prevalent religion throughout, so philisophical texts consistently address God from Christianity. Additionally, many of these are Western philosophies, so Christianity is especially relevant.

```python
sia = SentimentIntensityAnalyzer()

def avg_sentiment_context(school,word): #returns average sentiment for context around specified word 
    text = nltk.Text(concat_clean_words(school))
    conc_list = text.concordance_list(word,lines = 500)
    neg,neu,pos = 0,0,0
    n = len(conc_list)
    for i in range(n):
        sent_score = sia.polarity_scores(conc_list[i].line)
        neg += sent_score['neg']
        neu += sent_score['neu']
        pos += sent_score['pos']
    return {'neg': neg/n, 'neu': neu/n, 'pos': pos/n}
```

We will use 'God' as our topic of interest for each school.

```python
schools = ['feminism', 'capitalism', 'communism','nietzsche', 'analytic', 'continental']
neg_means = []
neu_means = []
pos_means = []
for school in schools:
    sent_dict = avg_sentiment_context(school,'God')
    neg_means.append(sent_dict['neg'])
    neu_means.append(sent_dict['neu'])
    pos_means.append(sent_dict['pos'])
```

We visualize the average sentiments using a bar graph. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(8,6))

x = np.arange(len(schools))
width = 0.2

rects1 = ax.bar(x - width, neg_means, width, label='negative')
rects2 = ax.bar(x, neu_means, width, label='neutral')
rects3 = ax.bar(x + width, pos_means, width, label='positive')

ax.set_ylabel('Sentiment Scores')
ax.set_title('Sentiment Scores surrounding God by school')
ax.set_xticks(x, schools)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)

fig.tight_layout()
plt.show()
```

Off first glance, the average sentiment surrounding God is neutral. This tends to be the case for philosophical texts, which observe and describe phenomena from a less personal and charged tone. However, it seems the majority of the selected schools also discuss God in a positive light, which has a 0 negativity value average. This is most especially true for Capitalism. A likely reason for this is a lower amount of data for capitalism than for other schools and their relative lack of interest in religion concerning their viewpoints.  Our baseline for negative Nietzshe expectedly had the largest average negativity value with 0.124. Regarding schools of modern relevance, feminism had the highest negativity value at 0.089. We can hypothesize that this is due to the patriarchal stereotypes that religion (and in this case, Christianity) insist onto society directly conflict with many values within feminism. Continental Philosophy had double the negativity value of the Analytical Philosophy (which had a much higher neutral value). This could be due to it being less interested in analysis of thought and knowledge.

To be sure, we use 'religion' below to see if there are any differences in the schools' language.

```python
neg_means = []
neu_means = []
pos_means = []
for school in schools:
    sent_dict = avg_sentiment_context(school,'religion')
    neg_means.append(sent_dict['neg'])
    neu_means.append(sent_dict['neu'])
    pos_means.append(sent_dict['pos'])
```

```python
fig, ax = plt.subplots(figsize=(8,6))

x = np.arange(len(schools))
width = 0.2

rects1 = ax.bar(x - width, neg_means, width, label='negative')
rects2 = ax.bar(x, neu_means, width, label='neutral')
rects3 = ax.bar(x + width, pos_means, width, label='positive')

ax.set_ylabel('Sentiment Scores')
ax.set_title('Sentiment Scores surrounding God by school')
ax.set_xticks(x, schools)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)

fig.tight_layout()
plt.show()
```

The neutral scores here are much higher in general. This could be because these texts discuss the word "religion" more often in more objective cases unrelated to their own schools of thought.

## What are the most common words these schools associate with God and religion?

Now we will create word clouds to visualize the most common words these texts associate with God. 

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
def word_cloud_synth(school,word): #synthesizes word cloud of context from data, inputting school and topic of interest
    text = nltk.Text(concat_clean_words(school))
    conc_list = text.concordance_list(word,lines = 500)
    conc_act_list = [conc_list[i].line for i in range(len(conc_list))]
    long_str = ' '.join(conc_act_list)
    return WordCloud().generate(long_str.replace('god','')) #remove God from cloud to showcase surrounding words
```

```python
plt.imshow(word_cloud_synth('feminism','God'), interpolation='bilinear')
plt.axis("off")
plt.show()
```

In relation to religion and Christian God, the most common words seen are "man", "woman", "love", "one", and "male". As a school which is very interested in the relationships between gender and society, we can see that this school also perceives Christian God through the lense of gender and sex. 

```python
plt.imshow(word_cloud_synth('capitalism','God'), interpolation='bilinear')
plt.axis("off")
plt.show()
```

Capitalism's most common words seem to be "agency", "want", "wonderful", "appearance", and "immediate". This is a sharp contrast to the gender lens that Feminism looked at through God. Here, there seems to be a stronger connection to humanity's free will and desires. Capitilism is about private parties owning and controlling property according to their desires. From this cloud, we can see these general values reflected in their contexts for religion. The more sparse word cloud confirms that there are less mentions of "God" in these texts compared to the other schools.

```python
plt.imshow(word_cloud_synth('communism','God'), interpolation='bilinear')
plt.axis("off")
plt.show()
```

Communism seems to connect God to the words "labour", "man", "law", "nature", and "system". Intrinsically, communism is interested in the labor of man and systematic government, so there is a connection between that and the perspectives for religion. As we know it, communism insists on a classless society and common, public ownership of mortal entities. This certainly reflects that. 

```python
plt.imshow(word_cloud_synth('nietzsche','God'), interpolation='bilinear')
plt.axis("off")
plt.show()
```

Nietzsche, the man who's philosophical thoughts are largely centered around the lack of faith in religion (and has a text called "God is Dead"), has the most common words associated as "one", "man", "life", "love", and "become". Man here most likely refers to people in general, and not as in gendered men. His views from this initial understanding are not so evident is this word cloud as they are for the others. However, they do showcase how he likely discussed the basis of love, life, and good in the context of God, whether in a pessimistic perspective or not. 

```python
plt.imshow(word_cloud_synth('analytic','God'), interpolation='bilinear')
plt.axis("off")
plt.show()
```

The most common words for Analytic Philosophy are "say", "may", "know", "exist", and "one". The core aspects of this field are the focuses on analytics of thought, knowledge, logic, and language. This is clearly reflected with words like "say" for language and "know" for knowledge and thought. This paints a picture of the texts looking at God through the lens of analytical language and thought. 

```python
plt.imshow(word_cloud_synth('continental','God'), interpolation='bilinear')
plt.axis("off")
plt.show()
```

The most common words for Continental Philosophy are "man", "one", "question", "death", "madness", and "saw". This is fairly different than Analytic Philosophy. This makes sense, as Continental Philosophy is often seen as the category which cover miscellaneous views which clearly differ from Analytic Philosophy. It's more interested in our perceptions and historical viewpoints, sometimes reject natural sciences. The more charged common words reflect this nature.
